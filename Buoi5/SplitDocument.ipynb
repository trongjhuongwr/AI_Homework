```python
text = "Lãnh thổ Việt Nam xuất hiện con người sinh sống từ thời đại đồ đá cũ, khởi đầu với các nhà nước Văn Lang, Âu Lạc. Âu Lạc bị nhà Triệu ở phương Bắc thôn tính vào đầu thế kỷ thứ 2 TCN sau đó là thời kỳ Bắc thuộc kéo dài hơn một thiên niên kỷ. Chế độ quân chủ độc lập được tái lập sau chiến thắng của Ngô Quyền trước nhà Nam Hán, mở đường cho các triều đại độc lập kế tục, nhiều lần chiến thắng trước các cuộc chiến tranh xâm lược từ phương Bắc đồng thời dần mở rộng về phía nam. Thời kỳ Bắc thuộc cuối cùng kết thúc sau chiến thắng trước nhà Minh của nghĩa quân Lam Sơn."
```

```python
def split_sentences(text):
    sentences = []
    current_sentence = ""
    
    for char in text:
        current_sentence += char
        
        if char in ['.', '!', '?']:
            next_index = text.index(char) + 1
            if next_index < len(text) and text[next_index].isspace():
                sentences.append(current_sentence.strip())
                current_sentence = ""
    
    if current_sentence:
        sentences.append(current_sentence.strip())
    
    return sentences

result = split_sentences(text)

for i, sentence in enumerate(result, 1):
    print(f"Câu {i}: {sentence}")
```

```python
def split_words(text):
    words = []
    current_word = ""
    
    for char in text:
        if char.isalnum():  # Ký tự chữ cái, số hoặc dấu nháy đơn
            current_word += char
        else:
            if current_word:
                words.append(current_word)
                current_word = ""
    
    if current_word:  # Thêm từ cuối cùng nếu có
        words.append(current_word)
    
    return words

result = split_words(text)

print("Các từ đã tách:")
for i, word in enumerate(result, 1):
    print(f"{i}. {word}")
```

```python
def count_words(text):
    # Tách từ
    words = split_words(text)

    # Đếm số lần xuất hiện của mỗi từ
    word_count = {}
    for word in words:
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1

    return word_count

result = count_words(text)

print("Thống kê số lần xuất hiện của các từ:")
for word, count in sorted(result.items(), key=lambda x: x[1], reverse=True):
    print(f"'{word}': {count}")
```

```python
# Định nghĩa thứ tự bảng chữ cái tiếng Việt
vietnamese_alphabet = 'aAàÀảẢãÃáÁạẠăĂằẰẳẲẵẴắẮặẶâÂầẦẩẨẫẪấẤậẬbBcCdDđĐeEèÈẻẺẽẼéÉẹẸêÊềỀểỂễỄếẾệỆfFgGhHiIìÌỉỈĩĨíÍịỊjJkKlLmMnNoOòÒỏỎõÕóÓọỌôÔồỒổỔỗỖốỐộỘơƠờỜởỞỡỠớỚợỢpPqQrRsStTuUùÙủỦũŨúÚụỤưƯừỪửỬữỮứỨựỰvVwWxXyYỳỲỷỶỹỸýÝỵỴzZ'

# Tách từ và đếm
words = []
current_word = ""
word_count = {}

for char in text:
    if char.isalnum() or char == "'":
        current_word += char.lower()
    else:
        if current_word:
            if current_word in word_count:
                word_count[current_word] += 1
            else:
                word_count[current_word] = 1
            words.append(current_word)
            current_word = ""

if current_word:  # Xử lý từ cuối cùng nếu có
    if current_word in word_count:
        word_count[current_word] += 1
    else:
        word_count[current_word] = 1
    words.append(current_word)

# Hàm so sánh cho việc sắp xếp
def compare_vietnamese(word1, word2):
    for c1, c2 in zip(word1, word2):
        i1 = vietnamese_alphabet.find(c1)
        i2 = vietnamese_alphabet.find(c2)
        if i1 == -1: i1 = len(vietnamese_alphabet)  # Xử lý ký tự không có trong bảng chữ cái
        if i2 == -1: i2 = len(vietnamese_alphabet)
        if i1 != i2:
            return i1 - i2
    return len(word1) - len(word2)

# Sắp xếp các từ theo bảng chữ cái tiếng Việt
words.sort(key=lambda word: [vietnamese_alphabet.find(c) if vietnamese_alphabet.find(c) != -1 else len(vietnamese_alphabet) for c in word])

# In kết quả
print("Thống kê số lần xuất hiện của các từ (sắp xếp theo bảng chữ cái tiếng Việt):")
printed_words = set()
for word in words:
    if word not in printed_words:
        print(f"'{word}': {word_count[word]}")
        printed_words.add(word)
```
